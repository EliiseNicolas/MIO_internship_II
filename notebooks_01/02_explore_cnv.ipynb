```python
cnv = pycnv.pycnv(path)

plt.figure(1, figsize=(12, 6))
plt.clf()

# Temperature profile
plt.subplot(1, 3, 1)
plt.plot(cnv.data['T0'], cnv.p)
plt.xlabel('Temperature [°C]')
plt.ylabel('Pressure [' + cnv.p_unit + ']')
plt.gca().invert_yaxis()
plt.title('Temperature')

# Conductivity profile
plt.subplot(1, 3, 2)
plt.plot(cnv.data['C0'], cnv.p)
plt.xlabel('Conductivity [S/m]')
plt.ylabel('Pressure [' + cnv.p_unit + ']')
plt.gca().invert_yaxis()
plt.title('Conductivity')

# Salinity profile
plt.subplot(1, 3, 3)
plt.plot(cnv.SA, cnv.p)
plt.xlabel('Absolute Salinity [' + cnv.SA_unit + ']')
plt.ylabel('Pressure [' + cnv.p_unit + ']')
plt.gca().invert_yaxis()
plt.title('Salinity')

plt.suptitle(f'CTD Cast — Lat: {cnv.lat:.3f}, Lon: {cnv.lon:.3f}')
plt.tight_layout()
plt.show()
```
